Iremos realizar a junção entre os DataFrames dados_fundos e dados_cadastro e trazendo somente as colunas que são de interesse do nosso estudo.

 Criamos a variável df_final através de um join trazendo os dados em comum das informações entre as duas tabelas.

Parametro how = "left" faz com que os arquivos sejam mantidos mesmo que não haja nome referente ao CNPJ procurado
caso o registro contenha na tabela "dados_cadastro" e bão tenha na tabela "dados_fundos", esse registro mão é juntado no df_final

````python


df_final = pd.merge(dados_fundos, dados_cadastro, how = "left",
                    left_on = ["CNPJ_FUNDO"],right_on=["CNPJ_FUNDO"])

df_final =  df_final[['CNPJ_FUNDO','DENOM_SOCIAL','DT_COMPTC','VL_QUOTA','VL_PATRIM_LIQ', 'NR_COTST','CLASSE','SIT']]
df_final
````

````python
import seaborn as sn
````

Verifcar quais as colunas possuem a melhor correlação entre elas.

````python
df_corresp =  df_final[['VL_QUOTA','VL_PATRIM_LIQ', 'NR_COTST']]
````

````python
correlacao = df_corresp.corr()

# plot da matriz de correlação

plot = sn.heatmap(correlacao, annot = True, fmt=".1f", linewidths=.6)
plot
````

Esta é a matriz gerada, a qual nos permitirá interpretar a relação entre as variávies. Nessa matriz podemos ver uma escala de cores de tonalidades de laranja, vermelho e roxo, e uma escala numérica que vai de 1 a -1. O interessante dessa representação gráfica é que além da barra lateral que demonstra a escala numérica do coeficiente e das cores, há também em cada quadrante o valor e a cor referente da relação. As variáveis que estiverem dentro da escala de 1 a 0,7 (+ ou -) e na tonalidade laranja no caso das positivas, e roxo escuro no caso das negativas, possuem uma forte correlação; já as variáveis que estão entre 0,7 a 0,5 (+ ou -) e na tonalidade avermelhada no caso das positivas e roxo médio no caso das negativas, possuem correlação moderada; as variáveis que possuem escala de 0,5 a 0,25 e (+ ou -) possuem baixa; e por fim as variáveis com coeficiente próximo a 0 (+ ou -) e com tonalidade vermelha não possuem correlação.
